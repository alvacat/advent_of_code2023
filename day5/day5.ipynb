```python
def match_source_to_dest(map_line):
    mappings={}
    source = int(map_line[1])
    destination = int(map_line[0])
    range_size = int(map_line[2])
    source_range = (source,source+range_size)
    destination_range = (destination,destination+range_size)
    mappings.update({source_range:destination_range})
    return mappings
```

```python
def make_mapping(mapping_dict,input_col_name):
    all_outputs = []
    for x in all_info_df.index:
        input_value = all_info_df[input_col_name][x]
        corresponding_output = -1

        for key in mapping_dict.keys():
            if input_value >= key[0] and input_value<= key[1]:
                position = input_value - key[0]
                outputs = mapping_dict[key]
                corresponding_output = outputs[0]+position

            if corresponding_output==-1:
                corresponding_output = input_value

        all_outputs.append(corresponding_output)
    return all_outputs
```
