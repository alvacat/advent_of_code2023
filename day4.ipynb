```python
import pandas as pd
```

```python
def split_line(line):
    inputs_and_outputs = line.split('|')
    return inputs_and_outputs

def find_inputs(inputs_and_outputs):
    inputs = inputs_and_outputs[0].split(' ')
    inputs = [int(x) for x in inputs if (not x.startswith('Card') and not x.endswith(':') and x!='')]
    return inputs

def find_outputs(inputs_and_outputs):
    outputs = inputs_and_outputs[1].split(' ')
    outputs = [int(x) for x in outputs if  x!='']
    return outputs

def check_outputs(inputs,outputs):
    good_outputs = []
    for output in outputs:
        if output in inputs:
            good_outputs.append(output)
    return good_outputs

def calculate_points(winning_numbers):
    number_of_winners = len(winning_numbers)
    if number_of_winners==0:
        points = 0
    else:
        points = 2**(number_of_winners - 1)
    return points
```

```python
df = pd.read_csv('input.txt',header=None)
df.columns = ['raw']
df.head()
```

```python
df['split_lines'] = df['raw'].apply(split_line)
df['input_numbers'] = df['split_lines'].apply(find_inputs)
df['output_numbers'] = df['split_lines'].apply(find_outputs)
df['good_outputs'] = df.apply(lambda x: check_outputs(x.input_numbers, x.output_numbers), axis=1)
df['points'] = df['good_outputs'].apply(calculate_points)
df.head()
```

```python
sum(df.points.values)
```

# Part 2

```python
copies_of_cards = {}

for x in df.index:
    num_winners = len(df['good_outputs'][x])
    times_extra_copies = 1
    if x in copies_of_cards:
        times_extra_copies += copies_of_cards[x]

    extra_cards = list(range(x+1,x+num_winners+1)) 
    for card_idx in extra_cards:
        if card_idx in copies_of_cards:
            copies_of_cards[card_idx]+=1 * times_extra_copies
        else:
            copies_of_cards[card_idx]=1 * times_extra_copies
```

```python
copies_col = []
for x in df.index:
    if x in copies_of_cards:
        copies_col.append(copies_of_cards[x])
    else:
        copies_col.append(0)

df['extra_copies'] = copies_col
df.head()
```

```python
df['total_copies'] = df['extra_copies']+1
df.head()
```

```python
df['total_points']=df['points']*df['total_copies']
```
