```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt',header=None)
df.columns = ['input_string']
```

```python
df.head()
```

```python
def find_number(input_string):
    for x in input_string:
        if x.isnumeric():
            first_number=x
            break
    return first_number

def find_both_numbers(input_string):
    first_num = find_number(input_string)
    backwards_string = ''
    x = len(input_string)-1
    while x>=0:
        backwards_string+=input_string[x]
        x-=1
    last_num = find_number(backwards_string)
    return first_num, last_num

def find_sum(input_string):
    first_num,last_num = find_both_numbers(input_string)
    return first_num+last_num
```

```python
df['no_numbers_flag']=df['input_string'].apply(str.isalpha)
```

```python
df.no_numbers_flag.value_counts()
```

```python
df['sum']=df.input_string.apply(find_sum)
```

```python
list_of_sums = list(df['sum'].values)
list_of_sums = [int(x) for x in list_of_sums]
```

```python
sum(list_of_sums)
```

```python
## Part 2

def change_to_numbers(input_string):
    input_string = input_string.replace('oneight','18')
    input_string = input_string.replace('twone','21')
    input_string = input_string.replace('threeight','38')
    input_string = input_string.replace('fiveight','58')
    input_string = input_string.replace('sevenine','79')
    input_string = input_string.replace('eightwo','82')
    input_string = input_string.replace('eighthree','83')
    input_string = input_string.replace('nineight','98')
    input_string = input_string.replace('zerone','01')
    input_string = input_string.replace('one','1')
    input_string = input_string.replace('two','2')
    input_string = input_string.replace('three','3')
    input_string = input_string.replace('four','4')
    input_string = input_string.replace('five','5')
    input_string = input_string.replace('six','6')
    input_string = input_string.replace('seven','7')
    input_string = input_string.replace('eight','8')
    input_string = input_string.replace('nine','9')
    input_string = input_string.replace('zero','0')
    return input_string
```

```python
df['processed_string']=df['input_string'].apply(change_to_numbers)
```

```python
df['processed_sum']=df.processed_string.apply(find_sum)
```
